```python
ifile = open('py4e.txt')
match = []
for line in ifile:
    line.rstrip()
    line_list = line.split()
    for item in line_list:
        try:
            a = int(item)
            match.append(a)
        except:
            continue
suma = 0
for num in match:
    suma += num
print(suma)
print(len(match))
```

```python
import re
ifile = open('py4e.txt')
abc = 0
for line in ifile:
    df = re.findall('[0-9]+',line)
for li in df:
    a = int(li)
    abc += a
print(abc)
```
